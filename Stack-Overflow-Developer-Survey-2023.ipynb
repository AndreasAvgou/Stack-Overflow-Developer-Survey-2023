```python
import pandas as pd

df = pd.read_csv('survey_results_public.csv')

# Grouping and saving various columns
df.groupby(['Country'])['MainBranch'].value_counts().reset_index(name='Counts').to_csv('csv/MainBranchCount.csv', header=True, index=False)
df.groupby(['Country'])['Age'].value_counts().reset_index(name='Counts').to_csv('csv/AgeCount.csv', header=True, index=False)
df.groupby(['Country'])['Employment'].value_counts().reset_index(name='Counts').to_csv('csv/EmploymentCount.csv', header=True, index=False)
df.groupby(['Country'])['RemoteWork'].value_counts().reset_index(name='Counts').to_csv('csv/RemoteWorkCount.csv', header=True, index=False)
df.groupby(['Country'])['EdLevel'].value_counts().reset_index(name='Counts').to_csv('csv/EdLevelCount.csv', header=True, index=False)
df.groupby(['Country'])['OrgSize'].value_counts().reset_index(name='Counts').to_csv('csv/OrgSizeCount.csv', header=True, index=False)
df.groupby(['Country'])['YearsCodePro'].value_counts().reset_index(name='Counts').to_csv('csv/YearsCodeProCount.csv', header=True, index=False)
df.groupby(['Country'])['DevType'].value_counts().reset_index(name='Counts').to_csv('csv/DevTypeCount.csv', header=True, index=False)
df.groupby(['Country'])['Industry'].value_counts().reset_index(name='Counts').to_csv('csv/IndustryCount.csv', header=True, index=False)

(df.groupby(['Country'])['MainBranch'].value_counts(normalize = True).mul(100).round(1).astype(str) + '%').reset_index(name='Percentege').to_csv('csv/MainBranch.csv',header=True, index = False)
(df.groupby(['Country'])['Age'].value_counts(normalize = True).mul(100).round(1).astype(str) + '%').reset_index(name='Percentege').to_csv('csv/Age.csv',header=True, index = False)
(df.groupby(['Country'])['Employment'].value_counts(normalize=True).mul(100).round(1).astype(str) + '%').reset_index(name='Percentege').to_csv('csv/Employment.csv',header=True, index = False)
(df.groupby(['Country'])['RemoteWork'].value_counts(normalize=True).mul(100).round(1).astype(str) + '%').reset_index(name='Percentege').to_csv('csv/RemoteWork.csv',header=True, index = False)
(df.groupby(['Country'])['OrgSize'].value_counts(normalize = True).mul(100).round(1).astype(str) + '%').reset_index(name='Percentege').to_csv('csv/OrgSize.csv',header=True, index = False)
(df.groupby(['Country'])['YearsCodePro'].value_counts(normalize = True).mul(100).round(1).astype(str) + '%').reset_index(name='Percentege').to_csv('csv/YearsCodePro.csv',header=True, index = False)
(df.groupby(['Country'])['EdLevel'].value_counts(normalize=True).mul(100).round(1).astype(str) + '%').reset_index(name='Percentege').to_csv('csv/EdLevel.csv',header=True, index = False)
(df.groupby(['Country'])['DevType'].value_counts(normalize=True).mul(100).round(1).astype(str) + '%').reset_index(name='Percentege').to_csv('csv/DevType.csv',header=True, index = False)
(df.groupby(['Country'])['Industry'].value_counts(normalize=True).mul(100).round(1).astype(str) + '%').reset_index(name='Percentege').to_csv('csv/Industry.csv',header=True, index = False)

# Reading various CSV files
csv1 = pd.read_csv("csv/MainBranch.csv")
csv2 = pd.read_csv("csv/MainBranchCount.csv")
csv3 = pd.read_csv("csv/Age.csv")
csv4 = pd.read_csv("csv/AgeCount.csv")
csv5 = pd.read_csv("csv/Employment.csv")
csv6 = pd.read_csv("csv/EmploymentCount.csv")
csv7 = pd.read_csv("csv/YearsCodePro.csv")
csv8 = pd.read_csv("csv/YearsCodeProCount.csv")
csv9 = pd.read_csv("csv/RemoteWork.csv")
csv10 = pd.read_csv("csv/RemoteWorkCount.csv")
csv11 = pd.read_csv("csv/EdLevel.csv")
csv12 = pd.read_csv("csv/EdLevelCount.csv")
csv13 = pd.read_csv("csv/OrgSize.csv")
csv14 = pd.read_csv("csv/OrgSizeCount.csv")
csv15 = pd.read_csv("csv/DevType.csv")
csv16 = pd.read_csv("csv/DevTypeCount.csv")
csv17 = pd.read_csv("csv/Industry.csv")
csv18 = pd.read_csv("csv/IndustryCount.csv")

# Merging and saving various grouped data
merged_data1 = csv2.merge(csv1, on=["Country", "MainBranch"]).reset_index(drop=True)
merged_data1.to_csv("MainBranchTotal.csv", index=False)

merged_data2 = csv4.merge(csv3, on=["Country", "Age"]).reset_index(drop=True)
merged_data2.to_csv("AgeTotal.csv", index=False)

merged_data3 = csv6.merge(csv5, on=["Country", "Employment"]).reset_index(drop=True)
merged_data3.to_csv("EmploymentTotal.csv", index=False)

merged_data4 = csv8.merge(csv7, on=["Country", "YearsCodePro"]).reset_index(drop=True)
merged_data4.to_csv("YearsCodeProTotal.csv", index=False)

merged_data5 = csv10.merge(csv9, on=["Country", "RemoteWork"]).reset_index(drop=True)
merged_data5.to_csv("RemoteWorkTotal.csv", index=False)

merged_data6 = csv12.merge(csv11, on=["Country", "EdLevel"]).reset_index(drop=True)
merged_data6.to_csv("EdLevelTotal.csv", index=False)

merged_data7 = csv14.merge(csv13, on=["Country", "OrgSize"]).reset_index(drop=True)
merged_data7.to_csv("OrgSizeTotal.csv", index=False)

merged_data8 = csv16.merge(csv15, on=["Country", "DevType"]).reset_index(drop=True)
merged_data8.to_csv("DevTypeTotal.csv", index=False)

merged_data9 = csv18.merge(csv17, on=["Country", "Industry"]).reset_index(drop=True)
merged_data9.to_csv("IndustryTotal.csv", index=False)

# Filtering and calculating percentages for Databases
Databases = ['MySQL', 'Microsoft SQL Server', 'Oracle', 'PostgreSQL', 'SQLite']
filterDatabases = df['DatabaseHaveWorkedWith'].isin(Databases)
grouped = df.loc[filterDatabases].groupby(['Country', 'DatabaseHaveWorkedWith']).size().reset_index(name='Counts')
grouped["Percentage"] = grouped.groupby("Country")["Counts"].apply(lambda x: x / x.sum() * 100)
grouped["Percentage"] = grouped["Percentage"].apply(lambda x: str(x) + '%')

print(grouped)
grouped.to_csv('DatabaseWorkedWith.csv', header=True, index=False)

# Filtering and calculating percentages for Languages
Languages = ['C', 'PHP', 'Python', 'Java', 'SQL']
filterLanguages = df['LanguageHaveWorkedWith'].isin(Languages)
grouped_l = df.loc[filterLanguages].groupby(['Country', 'LanguageHaveWorkedWith']).size().reset_index(name='Counts')
grouped_l["Percentage"] = grouped_l.groupby("Country")["Counts"].apply(lambda x: x / x.sum() * 100)
grouped_l["Percentage"] = grouped_l["Percentage"].apply(lambda x: str(x) + '%')

print(grouped_l)
grouped_l.to_csv('LanguageHaveWorkedWith.csv', header=True, index=False)
```
